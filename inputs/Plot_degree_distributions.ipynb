```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import networkx as nx 
```

```python
def get_degree_dist(nodestubs) :
    degree_dist = np.zeros(max(nodestubs)+1)
    for nodestub in nodestubs :
        
        degree_dist[nodestub]+=1
        
    return degree_dist
    
```

```python
# ER network..
ER_net = pd.read_csv("ER_network.csv")
nodestubs = np.concatenate((np.array(ER_net.Node1),np.array(ER_net.Node2)))

ER_degreedist = get_degree_dist(nodestubs)


# 0.001 rewired network..
rewire0001_net = pd.read_csv("ERtoPowerlaw_0.001rewired.csv")
nodestubs = np.concatenate((np.array(rewire0001_net.Node1),np.array(rewire0001_net.Node2)))
rewire0001_degreedist = get_degree_dist(nodestubs)

# 0.010 rewired network..
rewire0010_net = pd.read_csv("ERtoPowerlaw_0.010rewired.csv")
nodestubs = np.concatenate((np.array(rewire0010_net.Node1),np.array(rewire0010_net.Node2)))
rewire0010_degreedist = get_degree_dist(nodestubs)

# 0.100 rewired network..
rewire0100_net = pd.read_csv("ERtoPowerlaw_0.100rewired.csv")
nodestubs = np.concatenate((np.array(rewire0100_net.Node1),np.array(rewire0100_net.Node2)))
rewire0100_degreedist = get_degree_dist(nodestubs)

# 0.300 rewired network..
rewire0300_net = pd.read_csv("ERtoPowerlaw_0.300rewired.csv")
nodestubs = np.concatenate((np.array(rewire0300_net.Node1),np.array(rewire0300_net.Node2)))
rewire0300_degreedist = get_degree_dist(nodestubs)

# 0.500 rewired network..
rewire0500_net = pd.read_csv("ERtoPowerlaw_0.500rewired.csv")
nodestubs = np.concatenate((np.array(rewire0500_net.Node1),np.array(rewire0500_net.Node2)))
rewire0500_degreedist = get_degree_dist(nodestubs)

# 1.000 rewired network..
rewire1000_net = pd.read_csv("ERtoPowerlaw_1.000rewired.csv")
nodestubs = np.concatenate((np.array(rewire1000_net.Node1),np.array(rewire1000_net.Node2)))

rewire1000_degreedist = get_degree_dist(nodestubs)
```

```python
def get_xy_plots(degreedist) :
    x_vec = sorted(np.unique(degreedist))
    xtoy = dict()
    idx = 0
    for x in x_vec : xtoy[x_vec[idx]] = idx; idx +=1;

    y_vec = np.zeros(len(x_vec))
    for d in degreedist : y_vec[xtoy[d]] +=1;
    return x_vec, y_vec
```

```python
plt.figure(1,figsize=(3.6,3.6))


# ER
ER_degree_x, ER_degree_y = get_xy_plots(ER_degreedist)
plt.plot(ER_degree_x,ER_degree_y,color='C0',alpha=0.5,label='Rewire 0.000')

# Rewire0001 (1/1000)
rewire0001_x, rewire0001_y = get_xy_plots(rewire0001_degreedist)
plt.plot(rewire0001_x,rewire0001_y,color='C1',alpha=0.5,label='Rewire 0.001')

# Rewire0010 (1/100)
rewire0010_x, rewire0010_y = get_xy_plots(rewire0010_degreedist)
plt.plot(rewire0010_x,rewire0010_y,color='C2',alpha=0.5,label='Rewire 0.010')

# Rewire0100 (1/10)
rewire0100_x, rewire0100_y = get_xy_plots(rewire0100_degreedist)
plt.plot(rewire0100_x,rewire0100_y,color='C3',alpha=0.5,label='Rewire 0.100')

# Rewire0300 (3/10)
rewire0300_x, rewire0300_y = get_xy_plots(rewire0300_degreedist)
plt.plot(rewire0300_x,rewire0300_y,color='C4',alpha=0.5,label='Rewire 0.300')

# Rewire0500 (5/10)
rewire0500_x, rewire0500_y = get_xy_plots(rewire0500_degreedist)
plt.plot(rewire0500_x,rewire0500_y,color='C5',alpha=0.5,label='Rewire 0.500')


# Rewire1000 (all)
rewire1000_x, rewire1000_y = get_xy_plots(rewire1000_degreedist)
plt.plot(rewire1000_x,rewire1000_y,color='C6',alpha=0.5,label='Rewire 1.000')

#plt.hist(rewire1000_degreedist,color='C4',alpha=0.5)

plt.xscale("log")
plt.yscale("log")
plt.legend(frameon=False,fontsize=8)
plt.xlabel("Degree")
plt.ylabel("Multiplicity",)
plt.show()
```
